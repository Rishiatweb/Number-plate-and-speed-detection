```python
import os
import shutil
import random

image_dir = "/content/images"
train_dir = "/content/images/train"
val_dir = "/content/images/val"

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

all_images = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

random.shuffle(all_images)

split_idx = int(0.8 * len(all_images))
train_images = all_images[:split_idx]
val_images = all_images[split_idx:]

for img in train_images:
    shutil.move(os.path.join(image_dir, img), os.path.join(train_dir, img))

for img in val_images:
    shutil.move(os.path.join(image_dir, img), os.path.join(val_dir, img))

print(f"Moved {len(train_images)} images to train and {len(val_images)} images to val.")
```

```python
import numpy as np
import cv2
import easyocr
from ultralytics import YOLO
model = YOLO("yolov10n.pt")
reader = easyocr.Reader(['en'])
video_path = "/content/drive/MyDrive/traffic_video.mp4"
output_video_path = "processed_video1.avi"

cap = cv2.VideoCapture(video_path)
fps = int(cap.get(cv2.CAP_PROP_FPS))
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

vehicle_tracks = {}

pixel_to_kmh = 0.1

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    results = model(frame)
    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            cx, cy = (x1 + x2) // 2, (y1 + y2) // 2

            vehicle_id = None
            min_distance = float("inf")

            for vid, (prev_cx, prev_cy, prev_frame) in vehicle_tracks.items():
                distance = np.sqrt((cx - prev_cx) ** 2 + (cy - prev_cy) ** 2)
                if distance < 50:
                    if distance < min_distance:
                        min_distance = distance
                        vehicle_id = vid

            if vehicle_id is None:
                vehicle_id = len(vehicle_tracks) + 1

            speed = 0
            if vehicle_id in vehicle_tracks:
                prev_cx, prev_cy, prev_frame = vehicle_tracks[vehicle_id]
                frame_diff = max(1, cap.get(cv2.CAP_PROP_POS_FRAMES) - prev_frame)
                pixel_distance = np.sqrt((cx - prev_cx) ** 2 + (cy - prev_cy) ** 2)
                speed = (pixel_distance / frame_diff) * fps * pixel_to_kmh  # km/h

            vehicle_tracks[vehicle_id] = (cx, cy, cap.get(cv2.CAP_PROP_POS_FRAMES))

            plate = frame[y1:y2, x1:x2]
            plate_number = ""

            if plate.size > 0:
                gray_plate = cv2.cvtColor(plate, cv2.COLOR_BGR2GRAY)
                ocr_result = reader.readtext(gray_plate, detail=0)
                if ocr_result:
                    plate_number = ocr_result[0]

            cv2.putText(frame, f"ID {vehicle_id}: {plate_number} | {int(speed)} km/h",
                        (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

    out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()

print("Processing Complete. Output saved as 'processed_video1.avi'.")
```
